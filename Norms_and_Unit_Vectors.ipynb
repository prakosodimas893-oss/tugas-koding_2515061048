```python
# 1.buat vector x dengan ukuran 3x1
import numpy as np
x = np.array([48, 84, 7])
print(f"Vector x adalah {x}")
```

```python
#hitung nilai L2 norm dari vector x
np.linalg.norm(x)
print(f"Nilai L2 norm dari vector x adalah {x}")

```

```python
#2.buat vector v dengan elemen (3,4)
v = np.array([3, 4])
print(f"Vector v adalah {v}")
```

```python
#hitung nilai L2 norm dari vector v
np.linalg.norm(v)
print(f"Nilai L2 norm dari vector v adalah {v}")
```

```python
#tampilkan unit vector dari vector v
print(f"Unit vector dari vector v adalah {v}")
```

```python
#hitung panjang unit vector dari vector v
norm = np.linalg.norm(v)
print(f"panjang unit vector dari v adalah {norm}")
```
