```python
#1.buat matrix X berukuran 2x3 menggunakan numpy
import numpy as np
X = np.array([[1, 2],[3,4],[5,6]])
print(f"matrix X beukuran 2x3 adalah {X}")
```

```python
#tampilkan ukuran dari matrix X
X.shape
print(f"ukuran matrix X adalah {X.shape}")
```

```python
#tampilkan seluruh elemen dari kolom pertama dari matrix X
X[:,0]
print(f"elemen kolom pertama dari matrix X adalah {X[:,0]}")
```

```python
#tampilkan seluruh elemen dari baris kedua dari matrix X
X[:,1]
print(f"elemen baris kedua dari matrix X adalah {X[:,1]}")
```

```python
#2.ulangi soal nomor 1 menggunakan pytorch
import torch
#buat matrix X berukuran 2x3
X_pt = torch.tensor([[1, 2], [3, 4], [5, 6]])
print(f"matrix X berukuran  2x3 menggunakan pytorch adalah {X_pt}")
```

```python
#tampilkan ukuran dari matrix X
X_pt.shape
print(f"ukuran matrix X menggunakan pytorch adalah {X_pt.shape}")
```

```python
#tampilkan seluruh elemen dari kolom pertama dari matrix X
X_pt[:,0]
print(f"elemen kolom pertama dari matrix X menggunakan pytorch adalah {X_pt[:,0]}")
```

```python
#tampilkan seluruh elemen dari baris kedua dari matrix X
X_pt[:,1]
print(f"elemen baris kedua dari matrix X menggunakan pytorch adalah {X_pt[:,1]}")
```

```python
#ulangi soal nomor 1 menggunakan tensorflow
import tensorflow as tf
```

```python
#buat matrix X berukuran 2x3
X_tf = tf.Variable([[1, 2], [3, 4], [5, 6]])
print(f"matrix X berukuran  2x3 menggunakan tensorflow adalah {X_tf}")
```

```python
#tampilkan ukuran dari matrix X
X_tf.shape
print(f"ukuran matrix X menggunakan tensorflow adalah {X_tf.shape}")
```

```python
#tampilkan seluruh elemen dari kolom pertama dari matrix X
X_tf[:,0]
print(f"elemen kolom pertama dari matrix X menggunakan tensorflow adalah {X_tf[:,0]}")
```

```python
#tampilkan seluruh elemen dari baris kedua dari matrix X
X_tf[:,1]
print(f"elemen baris kedua dari matrix X menggunakan tensorflow adalah {X_tf[:,1]}")
```
