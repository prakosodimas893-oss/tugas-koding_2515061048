```python
# tampilkan tipe data dari y
dtype_tf = (y_tf)
print(f"tipe data saklar x adalah :{dtype_tf}")
```

```python
# jumlahkan x+y
x_tf + y_tf
print(f"hasil dari x+y adalah : {x_tf + y_tf}")
```

```python
# buat variabel bernama x_float dan simpan suatu nilai desimal
x_float = tf.Variable(84.0)
print(f"variabel x_float adalah : {x_float}")
```

```python
# jumlahkan x_float dengan y
tf_sum=tf.add(x_float, y_tf)
print(f"hasil dari x_float+y adalah : {tf_sum}")
```

```python
# tampilkan tipe data dari x_float
dtype_tf = (x_float)
print(f"tipe data saklar x adalah :{dtype_tf}")
```
