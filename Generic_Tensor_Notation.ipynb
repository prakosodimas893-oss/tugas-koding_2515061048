Sebagai contoh, tensor dengan rank 4 sering digunakan untuk merepresentasikan gambar (images), di mana setiap dimensi punya arti tertentu:

1. images_pt = torch.zeros([32, 28, 28, 3]) code ini di gunakan untuk membuat tensor yang berisi 0 dengan bentuk [32, 28, 28, 3]

2. 32 Sebagai jumlah gambar dalam satu batch training  (artinya ada 32 gambar diproses sekaligus)

3. 28 Sebagai tinggi gambar (height) dalam pixel (seperti dataset MNIST)

4. 28 Sebagai lebar gambar (width) dalam pixel   

5. 3 Sebagai jumlah channel warna (color channels)  untuk gambar berwarna (RGB = Red, Green, Blue)
