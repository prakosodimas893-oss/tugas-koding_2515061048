```python
#1.buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
import numpy as np
i = np.array([1, 0])
print(f"elemen vector i adalah {i}")
```

```python
j = np.array([0, 1])
print(f"elemen vector j adalah {j}")
```

```python
#buktikan bahwa vector i dan vector j itu orthogonal
np.dot(i, j)
print(f"hasil pembuktian orthogonal dari i dan j adalah {np.dot(i, j)}")
```

```python
#2.Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1, 2])
```

```python
w = np.array([2, -1])
```

```python
np.dot(v, w)
print(f"hasil pembuktian orthogonal dari v dan w adalah {np.dot(v, w)}")
```
